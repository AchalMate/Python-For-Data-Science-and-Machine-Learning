```python
count =0

class Hash_Linear_Probing:
    def __init__(self , size):
        self.Hash_Table = dict.fromkeys(range(size),-1)
        print(self.Hash_Table)  
        
    def hashing(self , key):
        return (key % size)
    
    def insert_into_hash_table_without_Replace(self , key , index):
        global count
        if self.Hash_Table[index] == -1:
            self.Hash_Table[index] = key
            count = count +1
        else:            
            self.find_vacant_index(key,index)
        
    def insert_into_hash_table_with_Replace(self , key , index):
      
        if self.Hash_Table[index] == -1:
            self.Hash_Table[index] = key
        elif index==self.hashing(self.Hash_Table[index]):
            self.find_vacant_index(key,index)
            
        else:
            temp = self.Hash_Table[index]
            self.Hash_Table[index]=key
            self.find_vacant_index(temp,index)
        
        
            
    def search_key(self ,key):
        if key in self.Hash_Table.values():
            print("Key Found :")
        else:
             print("Key Not Found :")
        self.display()
        
            
    def find_vacant_index(self , key , index):
        for i in range(1, size):
            pos = (index+i) % size
            if self.Hash_Table[pos] == -1:
                self.Hash_Table[pos] = key
                break; 
            
            
    
    def display(self):
#         print('Hash Table' , self.Hash_Table)
         
        for index,value in  self.Hash_Table.items():
            print(index,"==",value)
            
    
```
